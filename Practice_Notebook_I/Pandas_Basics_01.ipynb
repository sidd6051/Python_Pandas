```python
summer.head(10)
```

```python
summer.iloc[0,:3] # selecting rows and slicing columns
```

```python
summer.iloc[0,[0,1,2]] # selecting single row and selecting multiple columns
```

```python
summer.iloc[0,:3].equals(summer.iloc[0,[0,1,2]])
```

```python
summer.iloc[:,4].equals(summer.Country)
```

## Label-based Indexing and Slicing with loc[]

```python
summer
```

### selecting rows with loc[]

```python
summer.iloc[2] # selecting single row using index
```

```python
summer.loc['DRIVAS, Dimitrios'] # selecting single row using row label
```

```python
summer.loc['PHELPS, Michael']
```

### Indexing/Slicing Rows and Columns with loc[]

```python
summer.loc['PHELPS, Michael','Medal'] # selecting single row and column
```

```python
summer.loc['PHELPS, Michael',['Medal','Event']] # selecting single row and multiple columns
```

```python
summer.loc[:,['Medal','Event']] # Slicing rows and selecting multiple columns
```
