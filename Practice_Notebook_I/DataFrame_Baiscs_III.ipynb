## Ranking DataFrames with rank()

```python
sales = pd.Series([15,32,45,21,55,15,8], index = ['Mo','Tue','Wed','Thurs','Fri','Sat','Sun'])
```

```python
sales
```

```python
sales.sort_values(ascending = True)
```

```python
sales.rank(ascending = False, method = 'min').sort_values(ascending = True)
```

```python
sales.rank(ascending = True, method = 'min').sort_values(ascending = True)
```

```python
sales.rank(ascending = False,method = 'first').sort_values(ascending = True)
```

```python
titanic.head()
```

```python
titanic.fare.rank(ascending = False,method = 'min').sort_values().head(20)
```

```python
titanic['fare_rank'] = titanic.fare.rank(ascending = False,method = 'min')
```

```python
titanic.head()
```

```python
titanic.sort_values(by = 'fare_rank',ascending = True).head(30)
```

```python
titanic.drop(columns = 'fare_rank',inplace = True)
```

## apply(), map(), applymap()

```python
sales = pd.read_csv('sales_copy.csv',index_col = 0)
```

```python
sales.head()
```

```python
sales.info()
```

```python
sales.min(axis = 0)
```

```python
sales.max(axis = 'index')
```

```python
sales.min(axis = 'columns')
```

```python
sales.max(axis = 'columns')
```

```python
sales.apply(lambda series: series.max() - series.min(),axis = 0)
```

```python
sales.apply(lambda series : series.max() - series.min())
```

```python
sales.Mon.map(lambda ele : ele)
```

```python
sales.Mon.apply(lambda ele : ele)
```

```python
sales.applymap(lambda x:type(x))
```

```python
sales.apply(lambda x: len(x))
```

```python
sales.apply(lambda x: x.count(), axis = 'index')
```

## Hierarchical Indexing

```python
titanic.head()
```

```python
titanic = titanic.iloc[:50,:]
```

```python
titanic.set_index(['pclass','sex'],inplace = True)
```

```python
titanic.head(50)
```

```python
titanic.sort_index(ascending = [True,True])
```

```python
titanic.sort_index(ascending = [True,False],inplace=True)
```

```python
titanic
```

```python
titanic.swaplevel()
```

```python
titanic.reset_index(inplace=True)
```

```python
titanic
```

```python
titanic = titanic.set_index(['pclass','sex']).sort_index(ascending = True)
```

```python
titanic
```

```python
titanic.loc[(1,'female')]
```

```python
titanic.loc[(1,'female'),['age']]
```

```python
titanic.loc[(1,'female'),['age','fare']]
```

```python
titanic.loc[([1,2],'male'),:]
```
