# Joining on different column Lables & Indexes

```python
men2004 = pd.read_csv('men2004_copy.csv')
```

```python
men2008 = pd.read_csv('men2008_copy.csv')
```

```python
men2004.head()
```

```python
men2008.head()
```

```python
men2004.columns = ['Name','Medals']
```

```python
men2004.head()
```

```python
men0408 = men2004.merge(men2008, how = 'outer', left_on = 'Name', right_on = 'Athlete', 
                       suffixes = ('_2004','_2008'), indicator = True)
```

```python
men0408
```

```python
men0408.Name.fillna(men0408.Athlete, inplace = True)
```

```python
men0408
```

```python
men0408.drop(columns = ['Athlete', '_merge'], inplace = True)
```
