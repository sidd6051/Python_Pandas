## Importing Messy CSV Files

```python
pd.read_csv('titanic_raw_copy.csv')
```

```python
pd.read_csv('titanic_raw_copy.csv',skiprows = 3, skipfooter = 2, header = None,engine = 'python')
```

```python
column_names = ['Survived','Class','Gender','Age','SiSp','ParCh','fare','Emb','Deck']
```

```python
titanic = pd.read_csv('titanic_raw_copy.csv',skiprows = 3, skipfooter = 2, header = None,names = column_names) 
```

```python
titanic.head()
```

```python
titanic.to_csv('titanic_cleaned.csv',index = False)
```

```python
pd.read_csv('titanic_cleaned.csv')
```

```python
pd.read_excel('summer_raw_copy.xls',sheet_name = 1, skiprows = 2)
```

```python
pd.read_excel('summer_raw_copy.xls', sheet_name = 1, skiprows = [0,1],usecols = 'D:L')
```
