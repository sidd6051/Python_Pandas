# Pivoting DataFrames with pivot()

```python
table1 = pd.read_csv('table1_copy.csv')
```

```python
table1.head(10)
```

```python
table1.tail(10)
```

```python
table1.info()
```

```python
table1.shape
```

```python
table1.pivot(index = 'Country', columns = 'Medal', values = 'Count').fillna(0)
```

```python
table1.pivot(index = 'Medal',columns = 'Country', values = 'Count').fillna(0)#.unstack()
```

```python
table1_piv = table1.pivot(index = 'Country', columns = 'Medal', values = 'Count').fillna(0)
```

```python
table1_piv
```

```python
table1_piv.shape
```

```python
table1_piv.info()
```

```python
table1.head()
```

```python
table1.set_index(['Country','Medal']).unstack()
```

## Limits of Pivot()

```python
table2 = pd.read_csv('table2_copy.csv')
```

```python
table2.head(20)
```

```python
table2.tail(10)
```

```python
table2.info()
```

```python
table2.shape
```

```python
table2.pivot(index = 'Country', columns = 'Medal', values = 'Count')
```

```python
table2.groupby(['Country','Year','Medal']).Count.sum()
```

```python
table2_piv = table2.groupby(['Country','Year','Medal']).Count.sum().unstack()
```

```python
table2_piv
```

```python
table2_piv.info()
```

```python
table2_piv.shape
```
